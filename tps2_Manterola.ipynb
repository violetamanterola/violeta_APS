```python
# 2)
import numpy as np
import matplotlib.pyplot as plt

# datos generales, q como no se especificaron tomo valores arbitrarios. 
f0 = 1e3 #esta es la frecuencia central. La eleg√≠ yo, no especificaba ninguna.  
w0 = 2 * np.pi * f0 #frecuencia en radianes 

# expreso Q (no indica cual seria, elijo arbitrariamente pero puede cambairse) y el ancho de banda
q = 2 #factor de calidad o selectividad. Mientras mayor es el factor Q mas estrecho sera el ancho de banda!!
BW = w0 / q  # ancho de banda

def configuracion_plot(f,modulo_dB,f0,fase,nivel_3dB,f1,f2):
    #plt para el m√≥dulo de los filtros. 
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 1, 1)
    plt.semilogx(f, modulo_dB, color='b', label="|H(s)| [dB]")
    plt.axvline(f0, color='r', linestyle='--', label=f"f0 = {f0} Hz")  # marca en f0.
    plt.axhline(nivel_3dB, color='g', linestyle='--', label="-3 dB (desde m√°x)") #para indicar visualmente cuanto la gaanancia decae 3dB. 
    plt.xlabel("Frecuencia [Hz])")
    plt.ylabel("Ganancia [dB]")
    plt.title("Respuesta en Frecuencia (M√≥dulo)")
    plt.grid()
    plt.legend()
    # plt para la fase de los filtros. 
    plt.subplot(2, 1, 2)
    plt.semilogx(f, fase, color= 'm', label="‚à†H(s) [¬∞]")
    plt.axvline(f0, color='r', linestyle='--', label=f"f0 = {f0} Hz")
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Fase [¬∞]")
    plt.title("Respuesta en Frecuencia (Fase)")
    plt.grid()
    plt.legend()
```

```python
# A) caso filtro pasa banda
f = np.logspace(2, 4, 1500)  # rango de frecuencias para poder apreciar la banda. 
w = 2 * np.pi * f #para calcular H
s = 1j * w #para facilitar escribir la funcion de transferencia 
#funci√≥n de transferencia, calculada en el punto 1)
H = (w0 * s / q) / (s**2 + (w0 * s / q) + w0**2) 
# calculo m√≥dulo y fase
modulo_dB = 20 * np.log10(np.abs(H))  # M√≥dulo en decibeles
fase = np.angle(H, deg=True)  # Fase en grados
#calcular ca√≠da de 3dB
ganancia_max = np.max(modulo_dB) #calculo ganancia maxima para luego restarle los 3dB
nivel_3dB = ganancia_max - 3
configuracion_plot(f,modulo_dB,f0,fase,nivel_3dB,f1,f2)
plt.tight_layout()
plt.show()
```

```python
# B) caso filtro pasa altas segundo orden. 
f1 = np.logspace(1, 6, 1000) #rango de frecuencias.
w1 = 2 * np.pi * f1
s = 1j * w1 #para facilitar escribir la funcion de transferencia 
#funci√≥n de transferencia, calculada en el punto 1)
H = (s**2) / (s**2 + (w0 * s / q) + w0**2) 
# calculo modulo y fase
modulo_dB1 = 20 * np.log10(np.abs(H1))
fase1 = np.angle(H1, deg=True)

#calcular ca√≠da de 3dB
ganancia_max = np.max(modulo_dB1 )#calculo ganancia maxima para luego restarle los 3dB
nivel_3dB_1 = ganancia_max - 3

configuracion_plot(f1,modulo_dB1,f0,fase1,nivel_3dB_1 )
plt.tight_layout()
plt.show()
```

### *An√°lisis*
En primer lugar tenemos el filtro pasa bandas, que tiene este nombre ya que aten√∫a tanto bajas como altas frecuencias, permitiendo el paso solo de aquellas que se encuentren dentro de la banda de paso. Dicha banda se extiende hacia ambos lados de la frecuencia f0, que seria la de resonancia, la banda se extienede hasta la frecuencia de corte inferior y superior, donde la ganancia cae 3dB respecto a la m√°xima. Esto es lo que puede observarse en el gr√°fico del m√≥dulo, como pasiva todas las frecuencias menos aquellas dentro de la banda. Por su parte, la respuesta de fase, indica que es negativa para frecuencias menores, se acerca a 0 con la frecuencia de corte, y se vuelve positva para aquellas frecuencias mayores. 

En el caso del filtro pasa altos, la frecuencia de corte indica el inicio del ancho de banda, es aquella a los 3dB de ca√≠da de ganancia. Teniendo en cuenta que este filtro permite el paso de todas las frecuencias mayores a la de corte, podr√≠a decirse que su ancho de banda, *idealmente*, se extiende desde dicha frecuencia hasta el infinito. Como puede observarse en el gr√°fico, de la respuesta del modulo en frecuencia, aten√∫a las frecuencias menores a la frecuencia de corte y permite el paso del resto. En cuento a la respuesta de fase, se observa como la fase arranca aproximadamente desde 180¬∞, indicando una inversi√≥n de fase, y al lelgar a la frecuencia de corte comienza a descender notablemente, manteniendose en 0. 

Es importante destacar que mientras mayor sea Q (factor de calidad) mas estrecho sera el ancho de banda, indicando una mayor selectividad y menor p√©rdida de enrg√≠a. Adem√°s, influye en que tan pronunciado es el cambio de fase en los filtros, mientras mayor es el Q, m√°s pronunciada ser√°. (esto puede comprobarse variando el Q en el c√≥digo). Esto se ve expresado en la definici√≥n de ancho de banda: 
$$ BW =\frac{w0}{Q} $$
